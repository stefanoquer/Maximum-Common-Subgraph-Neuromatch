```python
import test_utility
import os

ts1 = []
for filename in sorted(os.listdir("/home/tom/tesi/MCSHeuristicCpp/test_results_total_similarity_H1/")):
    tt = test_utility.load_test("/home/tom/tesi/MCSHeuristicCpp/test_results_total_similarity_H1/" + filename)
    ts1.append(tt)

# ts2 = []
# for filename in sorted(os.listdir("/home/tom/tesi/MCSHeuristicCpp/precomputed_vectors_total_similarity_H1/")):
#     tt = test_utility.load_test("/home/tom/tesi/MCSHeuristicCpp/precomputed_vectors_total_similarity_H1/" + filename)
#     ts1.append(tt)

n1 = []
for filename in sorted(os.listdir("/home/tom/tesi/MCSHeuristicCpp/test_results_norm/")):
    tt = test_utility.load_test("/home/tom/tesi/MCSHeuristicCpp/test_results_norm/" + filename)
    n1.append(tt)


mc = []
for filename in sorted(os.listdir("/home/tom/tesi/MCSHeuristicCpp/test_results_classic/")):
    tt = test_utility.load_test("/home/tom/tesi/MCSHeuristicCpp/test_results_classic/" + filename)
    mc.append(tt)

def print_test(t1, t2):
   print("test name:", t1.test_description[0])
   print("test name2: ", t2.test_description[0])
   print("h1: ", t1.milestones[-1][0], "found in ", t1.milestones[-1][1], "recursions", " -- time: ", t1.total_time)
   print("mc: ", t2.milestones[-1][0], "found in ", t2.milestones[-1][1], "recursions", "-- time: ", t2.total_time)

def compare_h(heuristic1TestResult, heuristic2TestResult):
    h1Score = 0
    mcScore = 0
    tie = 0
    fails = []

    for t1 in heuristic1TestResult:
      for t2 in heuristic2TestResult:
        if t1.test_description[0] == t2.test_description[0]:
          print_test(t1, t2)
          if t1.milestones[-1][0] > t2.milestones[-1][0]:
            h1Score += 1
            # print("H won")
          elif t1.milestones[-1][0] < t2.milestones[-1][0]:
            mcScore += 1
            # print("Mc Won")
            fails.append(t2)
          elif t1.milestones[-1][1] > t2.milestones[-1][1]:
            mcScore += 1
            # print("Mc Won")
            fails.append(t2)
          elif t1.milestones[-1][1] < t2.milestones[-1][1]:
            h1Score += 1
            # print("H Won")
          else: 
            tie += 1
            fails.append(t2)

    # print("hscore: ", h1Score)
    # print("McSplit: ", mcScore)
    # print("Ties: ", tie)
    # print("")
    return fails, h1Score
```
